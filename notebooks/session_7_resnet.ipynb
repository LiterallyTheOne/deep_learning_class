# ResNet

```python
import keras
import numpy as np
```

* Residsidual network

* Skip connection

* Vanishing gradient
* Exploding gradient

```python
def identity_block(input_tensor, filter=64):
    x = keras.layers.Conv2D(
        filter,
        (1, 1),
        strides=(1, 1),
        padding="valid",
    )(input_tensor)
    x = keras.layers.BatchNormalization()(x)
    x = keras.layers.ReLU()(x)

    x = keras.layers.Conv2D(
        filter,
        (3, 3),
        strides=(1, 1),
        padding="same",
    )(x)
    x = keras.layers.BatchNormalization()(x)
    x = keras.layers.ReLU()(x)

    x = keras.layers.Conv2D(
        filter * 4,
        (1, 1),
        strides=(1, 1),
        padding="valid",
    )(x)
    x = keras.layers.BatchNormalization()(x)

    x = keras.layers.add([input_tensor, x])
    x = keras.layers.ReLU()(x)

    return x
```

```python
def convolutional_block(input_tensor, filter=64, strides=(2, 2)):
    x = keras.layers.Conv2D(
        filter,
        (1, 1),
        strides=strides,
        padding="valid",
    )(input_tensor)
    x = keras.layers.BatchNormalization()(x)
    x = keras.layers.ReLU()(x)

    x = keras.layers.Conv2D(
        filter,
        (3, 3),
        strides=(1, 1),
        padding="same",
    )(x)
    x = keras.layers.BatchNormalization()(x)
    x = keras.layers.ReLU()(x)

    x = keras.layers.Conv2D(
        filter * 4,
        (1, 1),
        strides=(1, 1),
        padding="valid",
    )(x)
    x = keras.layers.BatchNormalization()(x)

    shortcut = keras.layers.Conv2D(
        filter * 4,
        (1, 1),
        strides=strides,
        padding="valid",
    )(input_tensor)
    shortcut = keras.layers.BatchNormalization()(shortcut)

    x = keras.layers.Add()([shortcut, x])
    x = keras.layers.ReLU()(x)

    return x
```

```python
input_tensor = keras.layers.Input(shape=(224, 224, 3))

x = keras.layers.Conv2D(64, (7, 7), strides=(2, 2), padding="same")(input_tensor)
x = keras.layers.BatchNormalization()(x)
x = keras.layers.ReLU()(x)

x = keras.layers.MaxPooling2D((3, 3), strides=(2, 2), padding="same")(x)

x = convolutional_block(x, filter=64, strides=(1, 1))
x = identity_block(x, filter=64)
x = identity_block(x, filter=64)

x = convolutional_block(x, filter=128)
x = identity_block(x, filter=128)
x = identity_block(x, filter=128)
x = identity_block(x, filter=128)

x = convolutional_block(x, filter=256)
x = identity_block(x, filter=256)
x = identity_block(x, filter=256)
x = identity_block(x, filter=256)
x = identity_block(x, filter=256)
x = identity_block(x, filter=256)

x = convolutional_block(x, filter=512, strides=(1, 1))
x = identity_block(x, filter=512)
x = identity_block(x, filter=512)

x = keras.layers.GlobalAveragePooling2D()(x)
x = keras.layers.Dense(1000, activation="softmax")(x)

resnet_50 = keras.Model(input_tensor, x)
```
